```python
grouped_price = df.groupby('Dzielnica')['Cena'].median().sort_values(ascending=False)
df_grouped_price = pd.DataFrame(grouped_price)
df_grouped_price.reset_index(inplace=True)
df_grouped_price
```

```python
grouped_pricem2 = df.groupby('Dzielnica')['Cena za m2'].median().sort_values(ascending=False)
df_grouped_pricem2 = pd.DataFrame(grouped_pricem2)
df_grouped_pricem2.reset_index(inplace=True)
df_grouped_pricem2
```

```python
merged = df_grouped_price.merge(df_grouped_pricem2, on='Dzielnica', how='left')
merged.head()
```

```python
geolocation = Nominatim(user_agent="my_app", timeout=None)
merged["Wspolrzedne"]=merged["Dzielnica"].apply(geolocation.geocode)
merged["Szerokosc"] =  merged["Wspolrzedne"].apply(lambda x: x.latitude if x != None else None)
merged["Dlugosc"] =  merged["Wspolrzedne"].apply(lambda x: x.longitude if x != None else None)
```

```python
merged.drop(['Wspolrzedne'], axis=1, inplace=True)
merged.to_csv('clearing.csv', index=False)
```

```python
dfc = pd.read_csv('clearing.csv', delimiter=';', header=0)
```

```python

dfc= dfc.rename(columns={'Cena': 'Median cen', 'Cena za m2': 'Mediana cen za m2'})
dfc.head()
```

```python
fig = px.scatter_mapbox(dfc, lat='Szerokosc', lon='Dlugosc',title='Mediana m2 cen w dzielnicach: Marzec 2023', hover_name="Dzielnica", color='Mediana cen za m2', hover_data=["Median cen", "Mediana cen za m2"],
                        size="Mediana cen za m2", zoom=9.5, color_continuous_scale=px.colors.diverging.Geyser)
fig.update_layout(mapbox_style="carto-positron", autosize=True)
fig.show()
```
